```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Buat list predictions
predictions = [1] * 102
# Buat list true
true = [1] * 102

predictions[14] = 0
predictions[25] = 0
predictions[33] = 0
predictions[65] = 0
predictions[75] = 0
predictions[99] = 1

true[99] = 0
```

```python
# Menghitung akurasi
correct = sum(1 for p, t in zip(predictions, true) if p == t)
total = len(predictions)
accuracy = correct / total

print("Akurasi:", accuracy)
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
precision, recall, thresholds = precision_recall_curve(true, predictions)


plt.figure(figsize=(8, 6))
plt.plot(recall, precision, marker='.')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid(True)
plt.show()
```

```python
# Hitung confusion matrix
conf_matrix = confusion_matrix(true, predictions)

print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Membuat heatmap dari confusion matrix
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=['False', 'Motorcycle'], yticklabels=['False', 'Motorcycle'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Menghasilkan array indeks dari 0 hingga indeks terakhir
indices = np.arange(len(predictions))

# Membuat line chart
plt.figure(figsize=(16, 8))
plt.plot(indices, predictions, label='Predictions', marker='o', linestyle='-')
plt.plot(indices, true, label='True', marker='o', linestyle='-')

# Menambahkan label dan judul
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of Predictions vs True')
plt.legend()  # Menampilkan legenda
plt.grid(True)  # Menampilkan grid
plt.show()
```

```python
import matplotlib.pyplot as plt

# Contoh data loss dan metrik selama fine-tuning (ganti dengan data sesuai kebutuhan Anda)
epochs = range(1, 11)
train_loss = [0.5, 0.4, 0.3, 0.2, 0.15, 0.1, 0.09, 0.08, 0.07, 0.06]
val_loss = [0.45, 0.35, 0.25, 0.18, 0.14, 0.12, 0.1, 0.09, 0.08, 0.07]

# Membuat grafik loss
plt.figure(figsize=(8, 6))
plt.plot(epochs, train_loss, label='Training Loss', marker='o', linestyle='-')
plt.plot(epochs, val_loss, label='Validation Loss', marker='s', linestyle='-')

# Menambahkan label dan judul
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Fine-Tuning Loss')
plt.legend()  # Menampilkan legenda
plt.grid(True)  # Menampilkan grid

# Menampilkan grafik
plt.show()
```
